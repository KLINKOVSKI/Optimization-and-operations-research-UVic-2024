```python
# Print all subsets if there is at least one subset of set[]
# with a sum equal to the given sum
from sys import maxsize 
from itertools import permutations
import time
import matplotlib.pyplot as plt

# implementation of traveling Salesman Problem 
def subset_sum_time(set_, target_sum):
    start_time = time.time()
    
    flag = False

    def print_subset_sum(i, n, _set, target_sum, subset):
        global flag
        # If targetSum is zero, then there exists a subset
        if target_sum == 0:
            # Prints valid subset
            flag = True
            return

        if i == n:
            # Return if we have reached the end of the array
            return
            
        # Not considering the current element
        print_subset_sum(i + 1, n, _set, target_sum, subset)

        # Consider the current element if it is less than or equal to targetSum
        if _set[i] <= target_sum:
            # Push the current element into the subset
            subset.append(_set[i])
            
            # Recursive call for considering the current element
            print_subset_sum(i + 1, n, _set, target_sum - _set[i], subset)

            # Remove the last element after recursive call to restore subset's original configuration
            subset.pop()

    
    subset_ = []
    print_subset_sum(0, len(set_), set_, target_sum, subset_)
    end_time = time.time()
    return end_time - start_time

# Test cases
set_sizes = list(range(1, 21))
times = []

for size in set_sizes:
    test_set = list(range(1, size+1))
    target_sum = size * (size + 1) // 4  # Half of the sum of the set
    time_taken = subset_sum_time(test_set, target_sum)
    times.append(time_taken)

# Plotting
plt.plot(set_sizes, times, marker='o')
plt.xlabel('Number of Integers')
plt.ylabel('Time Taken (seconds)')
plt.title('Time Complexity of Subset Sum')
plt.grid(True)
plt.show()
```

```python
# Python3 program to implement traveling salesman 
# problem using naive approach. from sys import maxsize 
from itertools import permutations
import time
import matplotlib.pyplot as plt

# implementation of traveling Salesman Problem 
def travelling_salesman_time(graph):
    start_time = time.time()

    V = len(graph)

    def travellingSalesmanProblem(graph, s): 
        # store all vertex apart from source vertex 
        vertex = [] 
        for i in range(V): 
            if i != s: 
                vertex.append(i) 
        
        # store minimum weight Hamiltonian Cycle 
        min_path = maxsize 
        next_permutation = permutations(vertex)
        for i in next_permutation:

            # store current Path weight(cost) 
            current_pathweight = 0
            
            # compute current path weight 
            k = s 
            for j in i: 
                current_pathweight += graph[k][j] 
                k = j 
            current_pathweight += graph[k][s] 

            # update minimum 
            min_path = min(min_path, current_pathweight) 
        
        return min_path 

    s = 0
    result = travellingSalesmanProblem(graph, s)
    end_time = time.time()
    return end_time - start_time

# Test cases
graph_sizes = list(range(2, 13))
times = []

for size in graph_sizes:
    # Generating a simple graph with random weights for testing
    graph = [[0] * size for _ in range(size)]
    for i in range(size):
        for j in range(size):
            if i != j:
                graph[i][j] = (i + j) % 12 + 1  # Random weight between 1 and 12

    time_taken = travelling_salesman_time(graph)
    times.append(time_taken)

# Plotting
plt.plot(graph_sizes, times, marker='o')
plt.xlabel('Number of Cities')
plt.ylabel('Time Taken (seconds)')
plt.title('Time Complexity of Travelling Salesman')
plt.grid(True)
plt.show()
```
